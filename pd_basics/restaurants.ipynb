Отобрать записи, у которых более 20 категорий, сгруппировать по провинциям и вывести минимальную долготу в каждой, округлив ее до трех знаков после запятой. РЕзультат записать в файл .csv с разделителем ';'.

```python
res = df[df['categories'].str.split().apply(len) > 20].groupby('province', as_index = False).agg({'longitude': 'min'})
```

```python
res = round(res, 3)
```

```python
res
```

```python
res.to_csv('province_longitude.csv', sep=';')
```
